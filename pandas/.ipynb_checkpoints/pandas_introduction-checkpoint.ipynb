### Apply

>It applies function to dataframe or Series

```python
def square(x):
    return x * x
def f(x):
    return x + 40
def f2(x):
    return x -30
def apply(l, f):
    for i in range(len(l)):
        e = l[i]
        l[i] = f(e)
    
    return l    
l = [1,2,3,4]
apply(l, square)
```

```python
l
```

```python
apply(l, f)
```
